```python
#| export
# In this case we're using https://www.robots.ox.ac.uk/~vgg/data/pets/
# 'untar_data' is a thin wrapper for 'FastDownload.get'. It downloads and extracts 'url
path = untar_data(URLs.PETS)/'images'
```

```python
path
```

```python
path.ls()
```

```python
# Keep only the image files
get_image_files(path)
```

```python
#| export
# ImageDataLoaders - https://docs.fast.ai/vision.data.html#imagedataloaders
is_cat = lambda x: x[0].isupper()
dls = ImageDataLoaders.from_name_func('.',
    get_image_files(path), valid_pct=0.2, seed=42,
    label_func=is_cat,
    item_tfms=Resize(192))
```

```python
# We can then check if everything looks okay with the show_batch method (True is for cat, False is for dog):
dls.show_batch()
```
